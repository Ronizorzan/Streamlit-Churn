```python
from h2o import init, save_model, H2OFrame
from h2o.automl import H2OAutoML
from pandas import read_csv, DataFrame
import sklearn.metrics as metrics
from imblearn.under_sampling import TomekLinks
from imblearn.over_sampling import SMOTENC,SMOTE
```

```python
dados = read_csv("dados_tratados.csv")
dados.head()
```

```python
dados.isnull().all()
```

```python
dados.head(n=1)
```

```python
init(strict_version_check=False)
```

```python
imp = H2OFrame(dados)
treino, teste = imp.split_frame(ratios=[0.7], seed=321)
treino.head(1)
```

treino = treino.as_data_frame(use_multi_thread=True)
tomek = TomekLinks()
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(treino.drop("Churn", axis=1), treino["Churn"])
treino = DataFrame(X_resampled, columns=treino.drop("Churn", axis=1).columns)
treino["Churn"] = y_resampled
treino = H2OFrame(treino)
treino.shape

```python
treino["Churn"] = treino["Churn"].asfactor()
teste["Churn"] = teste["Churn"].asfactor()
```

```python
treino.shape
```

```python
modelo_h2o = H2OAutoML(nfolds=5, sort_metric="auto", max_runtime_secs=1800, seed=32, balance_classes=False, include_algos=["GBM", "StackedEnsemble"])
modelo_h2o.train(training_frame=treino, y="Churn", validation_frame=(teste))
```

```python
ranking = modelo_h2o.leaderboard
ranking = ranking.as_data_frame(use_multi_thread=True)
ranking
```

```python
ranking.tail()
```

```python
modelo_h2o.leader
```

```python
modelo_h2o.leader.model_performance(test_data=teste)
```

```python
previsoes = modelo_h2o.predict(test_data=teste)
previsoes = previsoes.as_data_frame(use_multi_thread=True)
previsoes
```

```python
teste["Churn"][:10]
```

```python
previsoes["predict"][:10]
```

```python
teste = teste.as_data_frame(use_multi_thread=True)
teste
```

```python
metrics.confusion_matrix( teste["Churn"], previsoes["predict"])
```

```python
save_model(model=modelo_h2o.leader, filename="Modelo_H2O", force=True, export_cross_validation_predictions=True)
```

```python
cm = metrics.ConfusionMatrixDisplay.from_predictions(teste["Churn"], previsoes["predict"])
```

```python
print(metrics.classification_report(teste["Churn"][:], previsoes["predict"][:]))
```
